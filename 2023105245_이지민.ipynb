### 4-1. 분석
#### 가설1: 담배값 인상 직전에 담배 판매량이 급격히 증가하고, 담배값 인상 이후 담배 판매량이 감소할 것이다. 그리고 이후에 서서히 담배 판매량은 증가할 것이다.
 2014년도와 2015년 그리고 2016년의 연간 담배 판매량에 중점을 두고 변화 양상을 확인해볼 것이다. 그리고 좀 더 자세하게 알기 위해서 2014년과 2015년 월별 담배 판매량 데이터를 시각화하여 변화 양상을 확인할 것이다. 

```python
df1
```

```python
#df1 표를 쉽게 그래프로 나타낼 수 있도록 행과 열을 바꿀 것이다. 
df1=df1.transpose()

#그래프를 만들기 용이하게 숫자 데이터만 남겨줄 것이다.
df1=df1.drop(["연도"],axis=0)

#담배 판매량 데이터를 object를 int로 바꿔줄 것이다. 
df1[0]=df1[0].astype(float)

df1.columns=['담배 판매량']

df1
```

```python
#좀 더 자세히 알아보기 위해서 2014년과 2015년 월별 담배 판매량 데이터를 시각화해서 알아볼 것이다. 

#먼저 데이터 뽑기에 방해가 되기 때문에 '14년, '15년 이라는 데이터 이름을 바꾸어 줄 것이다. 
df3=df3.rename(columns={"‘12년":"2012년","‘13년":"2013년","‘14년":"2014년","‘15년":"2015년"})

#맨 밑 월별 평균 데이터는 필요 없으므로 삭제하겠다. 
df3=df3.drop([12],axis=0)

#2014년 데이터와 2015년 데이터를 각각 추출할 것이다. 
df3_2014=df3.drop(["2012년","2013년","2015년"],axis=1)
df3_2014['2014년']=df3_2014['2014년'].astype(float)
df3_2015=df3.drop(["2012년","2013년","2014년"],axis=1)
df3_2015=df3_2015.head(7) #2015년의 데이터 중 8월 이후의 데이터는 기록되어 있지 않기 떄문에 7월까지의 데이터만 남길 것이다. 
df3_2015['2015년']=df3_2015['2015년'].astype(float)
df3_2015['2015년']=df3_2015['2015년'].astype(float)

#2014년 데이터와 2015년 데이터를 위아래로 합쳐서 그래프 그리기에 용이하도록 만들 것이다. 
df3_2014=df3_2014.rename(columns={"2014년":"담배 판매량"})
df3_2015=df3_2015.rename(columns={"2015년":"담배 판매량"})
df3=pd.concat([df3_2014, df3_2015], axis=0)
df3
```

```python
d1=df1.plot.bar()
plt.title("연도별 담배 판매량")
plt.xlabel("연도")
plt.ylabel("담배 판매량 (억 갑)")

d1.xaxis.set_major_locator(ticker.MultipleLocator(1))
d1.yaxis.set_major_locator(ticker.MultipleLocator(5))

d1

d3_=df3.plot.bar(x='구 분', y='담배 판매량', rot=0)
plt.title("2014년,2015년 월별 담배 판매량")
plt.xlabel("월")
plt.ylabel("담배 판매량")

d3_.xaxis.set_major_locator(ticker.MultipleLocator(1))
d3_.yaxis.set_major_locator(ticker.MultipleLocator(0.5))

d3_
```

- 첫번째 그래프에서 2014년, 2015년 그리고 2016년에 중점을 두고 확인해보면, 2015년 담배 판매량이 감소했다가 2016년 조금 증가한 것을 알 수 있다. 연간 담배 판매량을 보면 가설1은 '참'이다.

- 두번째 그래프를 통해서도 가설1은 '참'인 것을 알 수 있다.

### 4-2. 분석
#### 가설2: 전자담배 사용량이 증가할 것이다. 
 전자담배는 궐련형과 액상형이 있기 때문에 이 둘의 사용량 변화 추이를 알아볼 것이다. 

```python
df5
```

```python
df6
```

```python
#많은 데이터 중 성인 데이터만이 필요하기 때문에 이 데이터만 살려둘 것이다. 
df5=df5.head(2)

#데이터 구분을 위해 columns 이름을 바꿔줄 것이다. 
df5=df5.rename(columns={"Unnamed: 48":"2019년","Unnamed: 51":"2020년","Unnamed: 54":"2021년"})

#필요한 데이터만 추출하겠다. 
df5=df5[['2019년','2020년','2021년']]
df5=df5.drop(0,axis=0)

#나중에 액상형 전자담배 데이터와 합쳐야 하기 때문에 구분할 수 있도록 만들어준다. 
df5.insert(0, '구분', '궐련형(분율)')

#위 데이터와 같은 방식으로 필요한 데이터만 추출할 것이다. 
df6=df6.head(2)

df6=df6.rename(columns={"Unnamed: 30":"2013년","Unnamed: 33":"2014년","Unnamed: 36":"2015년","Unnamed: 39":"2016년","Unnamed: 42":"2017년","Unnamed: 45":"2018년","Unnamed: 48":"2019년","Unnamed: 51":"2020년","Unnamed: 54":"2021년"})

df6=df6[['2013년','2014년','2015년','2016년','2017년','2018년','2019년','2020년','2021년']]

df6=df6.drop(0,axis=0)

df6.insert(0, '구분', '액상형(분율)')

#궐련형 전자담배 같은 경우는 2019년 데이터부터 있기 때문에 2013년부터 2018년 데이터는 0으로 하고 삽입한다. 
df5.insert(1, '2013년', '0')
df5.insert(2, '2014년', '0')
df5.insert(3, '2015년', '0')
df5.insert(4, '2016년', '0')
df5.insert(5, '2017년', '0')
df5.insert(6, '2018년', '0')

#궐련형 데이터와 액상형 데이터를 위아래로 붙여준다. 
df5_6=pd.concat([df5,df6], axis=0)

#각 데이터의 type을 float로 바꿔준다.
df5_6['2013년']=df5_6['2013년'].astype(float)
df5_6['2014년']=df5_6['2014년'].astype(float)
df5_6['2015년']=df5_6['2015년'].astype(float)
df5_6['2016년']=df5_6['2016년'].astype(float)
df5_6['2017년']=df5_6['2017년'].astype(float)
df5_6['2018년']=df5_6['2018년'].astype(float)
df5_6['2019년']=df5_6['2019년'].astype(float)
df5_6['2020년']=df5_6['2020년'].astype(float)
df5_6['2021년']=df5_6['2021년'].astype(float)

#시각화하기에 용이하도록 행과 열을 바꿔줄 것이다. 
df5_6=df5_6.transpose()

df5_6.columns=['궐련형(분율)','액상형(분율)']

df5_6=df5_6.drop(['구분'],axis=0)

df5_6
```

```python
d5_6=df5_6.plot.bar(rot=0)
plt.title("전자담배 사용량")
plt.xlabel("년도")
plt.ylabel("전자담배 사용량")

d5_6.yaxis.set_major_locator(ticker.MultipleLocator(0.5))
plt.show()
```

### 4-2. 결론
#### 가설2: 전자담배 사용량이 증가할 것이다. 
- 궐련형 전자담배와 액상형 전자담배의 사용량이 증가하지 않았음을 볼 수 있다. 점점 증가하기 보다는 감소하는 것을 알 수 있다. 전자담배 사용량이 증가할 것이라는 가설은 '거짓'이다. 

### 4-3. 분석
#### 가설3: 전체적인 흡연율은 감소할 것이다. 
 흡연율 변화 추이를 살펴볼 것이다. 

```python
df4
```

```python
#필요한 데이터만 추출한다. 
df4=df4.drop([0,2,3],axis=0)
df4=df4.drop(['연령별(1)','성별(1)'],axis=1)

df4.index=['흡연율']

df4=df4.transpose()

df4['흡연율']=df4['흡연율'].astype(float)

df4
```

```python
d4_=df4.plot()
plt.title("흡연율 변화")
plt.xlabel("년도")
plt.ylabel("흡연율")
plt.ylim(0, 40)

d4_.yaxis.set_major_locator(ticker.MultipleLocator(1))
d4_.yaxis.set_major_locator(ticker.MultipleLocator(5))
plt.show()
```

 - 담배값이 증가함에 따라 전자담배 사용량이 늘 것이라 생각했지만 이 데이터만을 가지고는 알기 힘든 것 같다. 전자담배와 연초를 함께 피는 사람들도 있기에 이것에 대해서는 더 자세하고 다양한 데이터로 분석을 해야할 것 같다. 
 - 다가오는 2024년 담배값을 인상시킨다면 흡연율은 더욱 더 감소할 것으로 예상된다. 
 - 더 나아가 흡연율 감소에 영향을 미치는 다른 요인들을 알아보고, 이에 대해서도 분석할 필요가 있어보인다.
